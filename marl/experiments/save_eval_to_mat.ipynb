```python
import csv
import ast
import numpy as np
from scipy.io import savemat
# Load the CSV file
for ckp in [4,20,73,99,123]: #4,20,73,99,123; [4,40,70,100,135]
    csv_file = f"/home/kumquat/Documents/Linfan/social-agents-JAX/results/PopArtIMPALA_1_meltingpot_coop_mining_2025-03-14_14:39:21.229188/csv_logs/coop_mining{ckp}-timesteps.csv"
    matrices = []
    with open(csv_file, newline='\n', encoding='utf-8') as f:
        reader = csv.reader(f)
        next(reader)
        for row in reader:
            data = [ast.literal_eval(cell) for cell in row if cell]
            keys = data[0].keys()
            matrices.append({key: np.vstack([entry[key] for entry in data]) for key in keys})
    mat_dict = {"cell_array": np.array(matrices, dtype=object)}
    mat_dict = {"Ep": np.array(matrices, dtype=object)}
    savemat(csv_file[:-4]+'.mat',mat_dict)
```
